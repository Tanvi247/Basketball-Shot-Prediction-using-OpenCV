```python
import cv2
import cvzone
from cvzone.ColorModule import ColorFinder
import numpy as np
import math

cap = cv2.VideoCapture('Vid 6.mp4')

myColorFinder = ColorFinder(False)
hsvVals= {'hmin': 145, 'smin': 44, 'vmin': 110, 'hmax': 179, 'smax': 255, 'vmax': 255}

posListX, posListY = [], []
xList = [item for item in range(0,1800)]

while True:
    success,img=cap.read()
    #img = cv2.imread("Ball-1.png")
    #img = img[0:1500, :] 
    
    imgColor, mask = myColorFinder.update(img,hsvVals)
    imgContours, contours = cvzone.findContours(img,mask,minArea=500)
    
    if contours:
        posListX.append(contours[0]['center'][0])
        posListY.append(contours[0]['center'][1])
        
    if posListX:
        A,B,C = np.polyfit(posListX, posListY ,2)
        
        # Calculate the vertex of the parabola
        vertex_x = -B / (2 * A)
        vertex_y = A * vertex_x ** 2 + B * vertex_x + C
        
        # Draw the vertex
        cv2.circle(imgContours, (int(vertex_x), int(vertex_y)), 7, (255, 0, 0), cv2.FILLED)
        
        for i, (posX, posY) in enumerate(zip(posListX, posListY)):  
            pos = (posX,posY)
            cv2.circle(imgContours, pos,7, (0,255,0), cv2.FILLED)
            if i == 0:
                cv2.line(imgContours, pos, pos,(0,255,0),3)
            else:
                cv2.line(imgContours, pos, (posListX[i-1],posListY[i-1]),(0,255,0),3)
        
        if vertex_y >= 530:  # Proceed only if the vertex height is above 530
            a = A
            b = B
            c = C - 520

            discriminant = b ** 2 - (4 * a * c)
            if discriminant >= 0:  # Check if discriminant is non-negative
                x1 = int((-b + math.sqrt(discriminant)) / (2 * a))
                x2 = int((-b - math.sqrt(discriminant)) / (2 * a))
                
                print("Roots:", x1, x2)

                if 850 < x1 < 1010 or 850 < x2 < 1010:
                    print("Basket")
                else:
                    print("No Basket")
    
    imgContours = imgContours[0:1700, :]

    imgContours = cv2.resize(imgContours, (0, 0), None, 0.45, 0.45)
    #cv2.imshow("Image", img)
    cv2.imshow("ImageColor", imgContours)
    cv2.waitKey(50)
```
